# FEATURE ENGNEERING

```python
df['sex'] = df['sex'].map({'male': 0, 'female': 1}) #Run this cell to convert words to numbers.
df['smoker'] = df['smoker'].map({'yes': 1, 'no': 0})
df['region'] = df['region'].map({'southeast': 0, 'southwest': 1, 'northeast': 2, 'northwest': 3})
```

# FEATURE SELECTION

```python
x = df.drop(columns='charges',axis=1) #input features
y = df['charges'] #target variable
```

```python
x.shape
```

```python
y.shape
```

# FEATURE SCALING

```python
#DATA IN RANGE(STANDARDIZATION)
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

# MODEL SELECTION

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

x_train, x_test, y_train, y_test = train_test_split(x_scaled, y, test_size=0.2, random_state=2)
model = LinearRegression()
model.fit(x_train, y_train) # Ikkade model nerchukuntundi
```

# MODEL PREDICTION

```python
y_pred = model.predict(x_test)
from sklearn.metrics import r2_score
print("Accuracy Score:", r2_score(y_test, y_pred))
```

# SAVE MODEL

```python
import pickle
pickle.dump(model,open('insurance_model.pkl','wb'))
pickle.dump(scaler, open('scaler.pkl','wb'))
```
